```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Manually defined dataset (sorted alphabetically by 'LABEL')
data = {
    'LABEL': ['D', 'F', 'I', 'N', 'S'],
    'No': [1, 2, 0, 1, 5],    # Count of 'No' responses
    'Yes': [3, 2, 1, 2, 5]  # Count of 'Yes' responses
}

# Convert to DataFrame and sort alphabetically
df_counts = pd.DataFrame(data).set_index('LABEL').sort_index()

# Plot the horizontal stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))  # Increased height for better spacing
df_counts.plot(kind='barh', stacked=True, color=['red', 'blue'], ax=ax)

# Set x-axis ticks at 2-intervals
ax.set_xticks(np.arange(0, max(df_counts.sum(axis=1)) + 2, 2))

# Labels and title (formatted)
plt.xlabel("X-LABELS", fontsize=12, fontweight='bold', labelpad=10)
plt.ylabel("Y-LABELS", fontsize=12, fontweight='bold', labelpad=10)
plt.title("Bar Graph", fontsize=14, fontweight='bold', pad=25, loc='left')  # Increased padding

# Customizing the legend (higher above and centered)
legend = plt.legend(labels=["No", "Yes"], fontsize=10,
                    title_fontsize=12, loc='upper center', bbox_to_anchor=(0.3, 1.1), 
                    fancybox=True, shadow=True, ncol=2)
legend.get_frame().set_edgecolor("black")

# Add value labels at the tip inside the bars
for index, (yes, no) in enumerate(zip(df_counts["Yes"], df_counts["No"])):
    if no > 0:
        ax.text(no - 0.2, index, str(no), va='center', ha='right', fontsize=12, fontweight='bold', color='white')
    if yes > 0:
        ax.text(no + yes - 0.2, index, str(yes), va='center', ha='right', fontsize=12, fontweight='bold', color='white')

# Bold X and Y ticks
ax.tick_params(axis='both', labelsize=10, width=2)

# Adjust layout to prevent legend overlap
plt.tight_layout(rect=[0, 0.05, 1.2, 1])  # Reduced bottom margin

# Show plot
plt.show()
```
